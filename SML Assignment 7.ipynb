```r
# Create the formula to train Neural Network
formula <- paste(paste("Bankrupt.", sep=' + '),
                 paste(names(train)[2:96], collapse=' + '), sep=' ~ ')
substr(formula, start=1, stop=500)  # Print to check
```

NEURAL NETWORK

```r
# Train the neural network model
nn_md <- neuralnet(formula,          # The formula
                   train,                # Training data
                   hidden=c(1,30),
                   threshold = 0.603,# Size of the hidden layers
                   stepmax=10000,        # Maximum training step before stopping
                   rep=1,                 # Number of training repeat
                   lifesign='full',       # Print during train
                   algorithm='backprop',  # Algorithm to calculate the network (backpropagation)
                   learningrate=0.001,     # Learning rate
                   err.fct='ce',          # Error function, cross-entropy
                   act.fct="logistic",    # Function use to calculate the result (activation function)
                   linear.output=F        # FALSE = apply activation function to output layer
                  )
```

```r
# Make prediction and evaluation on train set
# Make prediction and evaluation on train

y_train_pred <- predict(nn_md, train[, c(2:96)])

mean((max.col(y_train_pred)-1) == train$Bankrupt.)
```

DEEP NEUREAL NETWORK

```r
# Train the neural network model
nn_md2 <- neuralnet(formula,          # The formula
                   train,                # Training data
                   hidden=c(12,24),
                   threshold = 0.1,# Size of the hidden layers
                   stepmax=10000,        # Maximum training step before stopping
                   rep=1,                 # Number of training repeat
                   lifesign='full',       # Print during train
                   algorithm='backprop',  # Algorithm to calculate the network (backpropagation)
                   learningrate=0.001,     # Learning rate
                   err.fct='ce',          # Error function, cross-entropy
                   act.fct="logistic",    # Function use to calculate the result (activation function)
                   linear.output=F        # FALSE = apply activation function to output layer
                  )
```

GLM MODEL

```r
glm.model <- glm(formula, data = train, family = binomial)
glm.model
```

```r
# Make predictions
library(dplyr)
probabilities <- glm.model %>% predict(test, type = "response")
predicted.classes <- ifelse(probabilities > 0.5, "pos", "neg")
# Model accuracy
mean(predicted.classes == test$Bankrupt.)
```

LDA

```r
library(MASS)
# Fit the model
lda.model <- lda(formula, data = train)
# Make predictions
predictions <- lda.model %>% predict(test)
# Model accuracy
mean(predictions$class==test$Bankrupt.)
```
